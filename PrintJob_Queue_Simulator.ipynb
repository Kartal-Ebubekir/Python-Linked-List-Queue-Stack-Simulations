```python
class Node:
  def __init__(self,fileName):
    self.fileName = fileName
    self.next = None
```

```python
class Queue():
  def __init__(self):
    self.front = None
    self.rear = None

  def enqueuePrintJob(self,fileName):
    newNode = Node(fileName)
    if self.rear is None :
      self.front = newNode
      self.rear = newNode
    else :
      self.rear.next = newNode
      self.rear = newNode

  def processNextJob(self):
    # If the list is empty
    if self.front is None :
      print("The queue is empty")
      return

    # If the list has just one item
    elif self.front.next is None :
      print(f"{self.front.fileName} proccess is done.There is no any proccess in the queue.")
      self.front = None
      self.rear = None
      return

    else :
      print(f"{self.front.fileName} proccess is done.Next proccess is {self.front.next.fileName}.")
      self.front = self.front.next


  def showQueue(self):
    current = self.front
    counter = 1
    print("Proccess list :")
    while current is not None:
      if self.front is None :
        print("The queue is empty")

      print(f"{counter}.'{current.fileName}' proccess")
      current = current.next
      counter += 1

    print("-------------------------------------")
```

```python
def main():
  q = Queue()
  q.enqueuePrintJob("Open the notebook")
  q.showQueue()
  q.enqueuePrintJob("Try some coding")
  q.showQueue()
  q.enqueuePrintJob("Save the notebook")
  q.showQueue()
  q.processNextJob()
  q.showQueue()
  q.processNextJob()
  q.showQueue()
  q.processNextJob()
  q.showQueue()
```

```python
if __name__ == "__main__" :
  main()
```
