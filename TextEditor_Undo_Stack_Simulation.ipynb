```python
class Node :
  def __init__(self,text) -> None:
    self.text = text
    self.next = None
```

```python
class undoSimulation:
  def __init__(self) -> None:
    self.head = None

  def pushWord(self,text):
    newNode = Node(text)
    if self.head is None :
      self.head = newNode
      return

    else :
      current = self.head
      while current.next is not None :
        current = current.next

      current.next = newNode

  def popWord(self):
    if self.head is None:
      pass

    elif self.head.next is None :
      self.head.text = ""
      return

    current = self.head
    while current.next.next is not None : # we go to item before the last item
      current = current.next

    current.next = None # Cutting connection

  def showWords(self):
    if self.head is None :
      print("")

    current = self.head
    print("- ",end="")
    while current is not None :
      print(f"{current.text}",end=" ")
      current = current.next
    print()
```

```python
def main():
  us = undoSimulation()
  us.pushWord("Hello")
  us.showWords()
  us.pushWord("World")
  us.showWords()
  us.popWord()
  us.showWords()
  us.popWord()
  us.showWords()
```

```python
if __name__ == "__main__":
  main()
```
