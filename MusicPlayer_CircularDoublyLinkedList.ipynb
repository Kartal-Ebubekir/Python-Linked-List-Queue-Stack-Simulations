```python

class Node:
  def __init__(self,name):
      self.name = name
      self.prev = None
      self.next = None
```

```python
class musicPlayer:
  def __init__(self):
    self.head = None
    self.tail = None
    self.currentSong = None # Currently playing song



  def addSongToEnd(self,name):
    newNode = Node(name)
    # If the list is empty , the new song will be head,tail and playing song
    if self.head is None :
      self.head = newNode
      self.tail = newNode
      self.currentSong = newNode
      print(f"'{name}' added the list (first song).")

      # The music player needs to be round, otherwise it won't continue after the songs finish.
      newNode.next = newNode
      newNode.prev = newNode
      return # for get out from the function


    # adding to the end
    newNode.next = self.head
    newNode.prev = self.tail
    self.tail.next = newNode
    self.head.prev = newNode
    self.tail = newNode


  def removeSong(self,name):

    if self.head is None :
      print("List is empty")
      return

    current = self.head
    flag = False
    while True :
      if name == current.name :
        flag = True
        break

      current = current.next

      # İf we came the first item again break the loop
      if current == self.head:
                break


    if not flag :
      print(f"Belitilen sarkı ({name}) listede bulunamadı.")
      return

    # if list has just one item
    if self.head == self.tail : # here  'and flag = True' check is unnecessary because if the flag is False it will already enter the above if block
      self.head = None
      self.tail = None
      self.currentSong = None
      return

    if current == self.head :
      self.head = self.head.next

    if current == self.tail :
      self.tail = self.tail.prev

    if current == self.currentSong:
      self.currentSong = self.currentSong.next
      print(f"The song {name} that was playing removed. Now {self.currentSong.name} playing.")

    else:
      print(f"{name} removed from list.")


    # We editting nodes.
    current.next.prev = current.prev
    current.prev.next = current.next


  def playNext(self):
    #if list is empty
    if self.head is None :
      print("List is empty")
      return

    self.currentSong = self.currentSong.next
    print(f"{self.currentSong.name} playing.\nNext song :{self.currentSong.next.name} , Previous song : {self.currentSong.prev.name}")

  def playPrevious(self):
    #if list is name
    if self.head is None :
      print("The list is empty")
      return

    self.currentSong = self.currentSong.prev
    print(f"{self.currentSong.name} playing.\nNext song :{self.currentSong.next.name} , Previous song : {self.currentSong.prev.name}")
```

```python
def main():
  mp = musicPlayer()

  # exm songs
  mp.addSongToEnd("Let it happen")
  mp.addSongToEnd("Sunflower")

  while True :
    print("--- Music Player Menu ---")
    print("1) Add song")
    print("2) Remove song")
    print("3) Next song")
    print("4) Previous song")
    print("5) Exit")

    choice = input("Choice : ")
    if choice == '1':
      name = input("Song name: ")
      mp.addSongToEnd(name)
    elif  choice == '2':
      isim = input("Song name to delete: ")
      mp.removeSong(name)
    elif  choice == '3':
      mp.playNext()
    elif  choice == '4':
      mp.playPrevious()
    elif  choice == '5':
      print("Closing...")
      break
    else:
      print("Geçersiz seçim.")
```

```python
if __name__ == "__main__":
    main()
```
